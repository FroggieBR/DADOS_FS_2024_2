M A T E M Á T I C A

```python
# Média de Pontuação em Matemática por Raça / Etnia
# df_mat = DataFrame 2 - Matemática
df_mat = df.groupby('race/ethnicity')['math score'].mean()
round(df_mat, 2)
```

```python
# Pode-se notar na tabela que, o grupo E está em primeiro lugar em pontuação, seguindo pelo D, C, B e A respectivamente.
```

```python
# Gráficos da Média de Pontuação de Matemática por Raça/Etnia
df_mat.plot(kind = 'barh', color = 'gray', title = 'Pontuação Média de Matemática por Raça/Etnia')
```

```python
# Seguindo a explicação acima do código do gráfico, mostra-se de forma visual o ranking de pontual média, onde o grupo E lidera
# e o grupo A fica em último.
```

Nível de Escolaridade por Média por Escrita

```python
df
```

```python
# Média de Pontuação em Escrita por Nível de Escolaridade dos Pais
# df_write = DataFrame 2 - Escrita
df_write = df.groupby('parental level of education')['writing score'].mean()
round(df_write, 2)
```

```python
# Pode-se notar que, os filhos que possuem os Pais com Mestrado e Bacharelado lideram o raking, tendo 75,68 e 73,38 respectivamente.
# Filhos de pais que Finalizaram o Ensino Médio, porém não frequentaram a Faculdade ficaram em ultimo lugar no ranking, com 62,45.
```

```python
# Gráficos da Média de Pontuação em Escrita por Nível de Escolaridade dos Pais
df_write.plot(kind = 'barh', color = 'gray', title = 'Pontuação Média em Escrita por Nível de Escolaridade dos Pais')
```

```python
# Apresentação gráfica da explicação sobre o Ranking da Pontuação Média em Escrita por Nível de Escolaridade dos Pais
```

Pontuação Média em Leitura por Participação em um Curso Preparatório para os Testes

```python
df
```

```python
# Média de Pontuação em Leitura por Participação em um Curso Preparatório para os Testes
# df_read = DataFrame 2 - Leitura
df_read = df.groupby('test preparation course')['reading score'].mean()
round(df_read, 2)
```

```python
# Pode-se mostrar que, pessoas que participaram de um Curso Preparatório tiveram notas maiores do que quem não participou.
```

```python
# Gráficos da Média de Pontuação em Leitura por Participação em um Curso Preparatório para os Testes
df_read.plot(kind = 'barh', color = 'gray', title = 'Pontuação Média em Leitura por Participação em Curso Preparatório')
```

```python
# Nesse visual de gráficos de barras apresenta a superioridade em notas de quem participou de um curso preparatório com quem não participou.
```
